```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import glob as glob
plt.rcParams.update({'figure.max_open_warning': 0})
%matplotlib inline
```

```python
n = np.random.randint(1,100)


train = h5py.File(r'D:\Landslide\TrainData\img\image_'+str(n)+'.h5', 'r')
print(train.keys())

label = h5py.File(r'D:\Landslide\TrainData\mask\mask_'+str(n)+'.h5','r')
print(label.keys())

data = np.array(train.get('img'))
label_data = np.array(label.get('mask'))

print(data.shape,label_data.shape)

#Training image
for i in range(14):
    plt.figure();
    fig,(ax1,ax2)= plt.subplots(1,2,figsize=(10,10))
    ax1.imshow(data[:,:,i]);
    ax1.set_title('Training data');
    ax2.imshow(label_data);
    ax2.set_title('Label where Landslides occured')
```
