```python
import os
import numpy as np
import matplotlib.pyplot as plt
import elevation
import richdem as rd
from numpy import asarray
```

```python
dem_data = '/content/swisssurface3d-raster_2018_2570-1163_0.5_2056_5728.tif'
```

```python
dem_data = rd.LoadGDAL(dem_data)
```

```python
type(dem_data)
```

```python
plt.imshow(dem_data)
```

```python
slope = rd.TerrainAttribute(dem_data, attrib='slope_riserun')
rd.rdShow(slope,cmap='jet' ,axes=False, figsize=(8, 5.5))
plt.show()
```

```python
aspect = rd.TerrainAttribute(dem_data, attrib='aspect')
rd.rdShow(aspect, axes=False, cmap='jet', figsize=(8, 5.5))
plt.show()
```

```python
def calculate_slope(dem_data):
  slope = rd.TerrainAttribute(dem_data, attrib='slope_degrees')
  aspect = rd.TerrainAttribute(dem_data, attrib='aspect')
  return slope,aspect
```

```python
slope1,aspect1 = calculate_slope(dem_data)
```

```python
rd.rdShow(slope1,axes=False,cmap='jet',figsize=(8, 5.5))
```

```python
rd.rdShow(aspect1,axes=False,cmap='jet',figsize=(8,5.5))
```
